## Q1

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("datasets/netflix_data.csv")
data.head()
```

```python
data.tail()
```

```python
def cast_len(val):
    if val == "None":
        return 0
    else:
        return len(str(val).split(", "))
```

```python
data["cast"].isnull().sum()
```

```python
data["cast"].replace(np.nan, "None",inplace=True)
```

```python
data.tail()
```

```python
data["Cast_Length"] = data["cast"].apply(cast_len)
```

```python
data.tail()
```

```python
x = np.linspace(0,10,1000)
plt.plot(x, np.sin(x), color='#000')
plt.plot(x, np.sin(x-1), color = 'g')
plt.plot(x, np.sin(x-2) ,color = 'r')
plt.plot(x, np.sin(x-3) ,color = 'b')
plt.plot(x, np.sin(x-4) ,color = 'chartreuse')

plt.show()
```

```python
x = np.linspace(0,10,1000)

plt.plot(x, np.sin(x), color='#000', linestyle="solid", label="xyzkshd")
plt.plot(x, np.sin(x-1), color = 'g', linestyle="dashed", label="sin(x-1)")
plt.plot(x, np.sin(x-2) ,color = 'r', linestyle="dotted", label="sin(x-2)")
plt.plot(x, np.sin(x-3) ,color = 'b', linestyle="dashdot", label="sin(x-3)" )

plt.title("Sin Graph")
plt.xlabel("x")
plt.ylabel("sin(x)")

plt.legend()
plt.show()
```
