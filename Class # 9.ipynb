### Scatter Plot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)

plt.plot(x, y, 'o', color='black')
```

```python
plt.plot(x, y, '-ok', color='black')
```

```python
plt.scatter(x,y)
```

```python
data = np.random.randn(1000)
# print(data)
plt.hist(data, bins=10, histtype="stepfilled")
plt.show()
```

```python
x = np.linspace(0,10,1000)
inp = input("Enter function Sin or Cos")
if inp=="Sin":
    plt.plot(x, np.sin(x), label="Sine")
else:  
    plt.plot(x, np.cos(x), label="Cosine")

plt.legend()
```

```python

plt.plot(x, np.sin(x), label="Sine")
plt.plot(x, np.cos(x), label="Cosine")

plt.legend(loc="lower left", ncol=2, fancybox=True,borderpad=1, shadow = True)
```
